```python
# ============================================
# LOAD THE STUDENT DROPOUT DATASET
# ============================================

# File path
file_path = "../DATA/student_dropout_dataset_v3.xlsx - Clean DataSet.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first five rows
df.head()
```

```python
#Is there a statistically significant relationship between travel time and dropout?

# Calculate the average travel time by dropout status
df.groupby("Dropout")["Travel_Time_Minutes"].mean()
```

```python
#Discussion

#Which group has the higher average travel time? group 1 (students who dropped out) has a higher average travel time compared to group 0 (students who did not drop out).

#Is the difference large enough to suggest that commuting could be a contributing factor to dropout? it appears that students who dropped out have a higher average travel time, which may suggest that longer commutes could be a contributing factor to dropout. However, further statistical analysis would be needed to confirm this relationship.

#SDG Link: This analysis relates to SDG 4 (Quality Education) because long travel times can reduce access to education, and SDG 11 (Sustainable Cities and Communities) because transport and proximity to educational institutions influence students' ability to attend consistently.
```

```python
#DISTRIBUTION TABLE

plt.figure(figsize=(8,5))

sns.histplot(data=df,
             x="Stress_Index",
             hue="Gender",
             bins=10,
             kde=True)

plt.title("Distribution of Stress Index by Gender")

plt.xlabel("Stress Index")

plt.ylabel("Number of Students")

plt.show()
```

```python
#Explanation
#This chart compares the distribution of stress scores for male and female students.

#From the histogram, we can observe that the female gender tends to have a higher concentration of stress scores at the upper end of the scale, indicating that female students may experience higher stress levels compared to male students.
```

```python
#INTERPRETATION

#THIS BOXPLOT SHOWS THAT FEMALE STUDENTS HAVE A HIGHER MEDIAN STRESS INDEX AND A WIDER RANGE OF STRESS SCORES COMPARED TO MALE STUDENTS.
```

```python
#AVERAGE STRESS INDEX BY GENDER AND DROPOUT STATUS

# Calculate average stress by gender and dropout status
df.groupby(["Gender", "Dropout"])["Stress_Index"].mean()
```

```python
#IN CONCLUSION
#Female students recorded a higher average Stress Index than male students. 
#This may suggest that female students in this dataset experience greater academic or personal stress. 
#However, the dataset does not contain variables explaining the causes of stress, so the observed difference should not be interpreted as evidence of causation. 
#Factors such as academic workload, financial pressure, family responsibilities, or access to support services may contribute and would require further investigation.


#SDG Connection
#SDG 3 – Good Health and Well-being: Understanding stress levels can help institutions design mental health and counseling interventions.
#SDG 4 – Quality Education: Reducing stress may improve student engagement, academic performance, and retention.
#SDG 5 – Gender Equality: If meaningful gender differences are observed, universities can evaluate whether support services adequately address the needs of all students.
```
